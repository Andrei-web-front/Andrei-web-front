```python
import numpy as np

def floyd_warshall(graph):
    num_vertices = len(graph)
    dist = np.array(graph)

    for k in range(num_vertices):
        for i in range(num_vertices):
            for j in range(num_vertices):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]

    return dist

# Пример использования
graph = [
    [0, 3, float('inf'), 5],
    [2, 0, float('inf'), 4],
    [float('inf'), 1, 0, float('inf')],
    [float('inf'), float('inf'), 2, 0]
]

shortest_paths = floyd_warshall(graph)
print(shortest_paths)
```

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from itertools import combinations

# Данные
cities = ["Москва", "Санкт-Петербург", "Казань", "Екатеринбург",
          "Новосибирск", "Ростов-на-Дону", "Владивосток", "Сочи",
          "Калининград", "Краснодар"]

dist_matrix = np.array([
    [0, 700, 800, 1800, 3300, 1100, 9200, 1600, 1250, 1350],
    [700, 0, 1500, 2500, 3800, 1900, 9900, 2300, 800, 2200],
    [800, 1500, 0, 900, 2500, 1400, 8400, 1900, 2000, 1600],
    [1800, 2500, 900, 0, 1600, 2100, 7500, 2800, 2900, 2400],
    [3300, 3800, 2500, 1600, 0, 3700, 5900, 4400, 4500, 4000],
    [1100, 1900, 1400, 2100, 3700, 0, 10100, 500, 2350, 300],
    [9200, 9900, 8400, 7500, 5900, 10100, 0, 10600, 10450, 10400],
    [1600, 2300, 1900, 2800, 4400, 500, 10600, 0, 2850, 350],
    [1250, 800, 2000, 2900, 4500, 2350, 10450, 2850, 0, 2600],
    [1350, 2200, 1600, 2400, 4000, 300, 10400, 350, 2600, 0]
])

# Функция для восстановления пути
def reconstruct_path(path_matrix, start, end):
    path = [end]
    while path[-1] != start:
        path.append(path_matrix[start][path[-1]])
    return path[::-1]

# Алгоритм Флойда-Уоршелла с сохранением путей
def floyd_warshall_with_paths(graph):
    n = len(graph)
    dist = graph.copy()
    next_node = np.zeros((n, n), dtype=int)

    for i in range(n):
        for j in range(n):
            next_node[i][j] = j if i != j and graph[i][j] != 999 else -1

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]
                    next_node[i][j] = next_node[i][k]

    return dist, next_node

# Вычисляем кратчайшие пути
shortest_dists, path_matrix = floyd_warshall_with_paths(dist_matrix)

# Функция для визуализации
def visualize_route(start_city, end_city):
    start_idx = cities.index(start_city)
    end_idx = cities.index(end_city)

    # Получаем путь
    route = reconstruct_path(path_matrix, start_idx, end_idx)
    route_cities = [cities[i] for i in route]
    distance = shortest_dists[start_idx][end_idx]

    print(f"Оптимальный маршрут: {' → '.join(route_cities)}")
    print(f"Общее расстояние: {distance} км")

    # Создаём граф
    G = nx.Graph()
    for city in cities:
        G.add_node(city)

    # Добавляем рёбра
    for i, j in combinations(range(len(cities)), 2):
        if dist_matrix[i][j] < 9999:
            G.add_edge(cities[i], cities[j], weight=dist_matrix[i][j])

    # Позиционирование узлов
    pos = nx.spring_layout(G, seed=42)

    # Рисуем граф
    plt.figure(figsize=(12, 8))
    nx.draw_networkx_nodes(G, pos, node_size=700, node_color='lightblue')
    nx.draw_networkx_edges(G, pos, width=1, alpha=0.3, edge_color='gray')

    # Выделяем маршрут
    route_edges = [(route_cities[i], route_cities[i+1]) for i in range(len(route_cities)-1)]
    nx.draw_networkx_edges(G, pos, edgelist=route_edges, width=3, edge_color='red')

    nx.draw_networkx_labels(G, pos, font_size=10)

    # Подписываем рёбра маршрута
    edge_labels = {(u, v): dist_matrix[cities.index(u)][cities.index(v)]
                  for u, v in route_edges}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)

    plt.title(f"Маршрут: {start_city} → {end_city} ({distance} км)", fontsize=14)
    plt.axis('off')
    plt.tight_layout()
    plt.show()

# Интерфейс для выбора городов
print("Доступные города:")
for i, city in enumerate(cities, 1):
    print(f"{i}. {city}")

while True:
    try:
        start_num = int(input("\nВведите номер начального города: ")) - 1
        end_num = int(input("Введите номер конечного города: ")) - 1

        if 0 <= start_num < len(cities) and 0 <= end_num < len(cities):
            visualize_route(cities[start_num], cities[end_num])
        else:
            print("Ошибка: введите числа от 1 до", len(cities))

        if input("\nПостроить ещё маршрут? (y/n): ").lower() != 'y':
            break
    except ValueError:
        print("Ошибка: введите числа!")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
from itertools import combinations

# Упрощённые данные для Colab (меньше городов)
cities = ["Москва", "Санкт-Петербург", "Казань", "Сочи", "Калининград"]
coords = {
    "Москва": (37.6176, 55.7558),
    "Санкт-Петербург": (30.3351, 59.9343),
    "Казань": (49.1088, 55.7961),
    "Сочи": (39.7233, 43.5855),
    "Калининград": (20.4542, 54.7104)
}

# Матрица расстояний (км)
dist_matrix = np.array([
    [0, 700, 800, 1600, 1250],
    [700, 0, 1500, 2300, 800],
    [800, 1500, 0, 1900, 2000],
    [1600, 2300, 1900, 0, 2850],
    [1250, 800, 2000, 2850, 0]
])

# Упрощённый алгоритм
def floyd_warshall(graph):
    n = len(graph)
    dist = graph.copy()
    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]
    return dist

shortest_dists = floyd_warshall(dist_matrix)

# Визуализация
def show_route(start_city, end_city):
    try:
        # Загружаем упрощённую карту
        world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
        russia = world[world.name == 'Russia']

        fig, ax = plt.subplots(figsize=(10, 6))
        russia.plot(ax=ax, color='lightgray', edgecolor='black')

        # Рисуем города
        for city, (lon, lat) in coords.items():
            ax.plot(lon, lat, 'o', markersize=8,
                   color='red' if city in [start_city, end_city] else 'blue')
            ax.text(lon+1, lat, city, fontsize=9)

        # Рисуем маршрут (упрощённо)
        start_lon, start_lat = coords[start_city]
        end_lon, end_lat = coords[end_city]
        ax.plot([start_lon, end_lon], [start_lat, end_lat],
               'r--', linewidth=2)

        # Расстояние
        start_idx = cities.index(start_city)
        end_idx = cities.index(end_city)
        distance = shortest_dists[start_idx][end_idx]

        plt.title(f"{start_city} → {end_city} (~{distance} км)", fontsize=12)
        plt.grid()
        plt.show()

    except Exception as e:
        print(f"Ошибка визуализации: {e}. Показываю текстовый результат:")
        start_idx = cities.index(start_city)
        end_idx = cities.index(end_city)
        print(f"Кратчайшее расстояние: {shortest_dists[start_idx][end_idx]} км")

# Интерфейс
print("Доступные города:")
for i, city in enumerate(cities, 1):
    print(f"{i}. {city}")

while True:
    try:
        start_num = int(input("\nНомер начального города (1-5): ")) - 1
        end_num = int(input("Номер конечного города (1-5): ")) - 1

        if 0 <= start_num < 5 and 0 <= end_num < 5:
            show_route(cities[start_num], cities[end_num])
        else:
            print("Ошибка: введите числа от 1 до 5")

        if input("Продолжить? (y/n): ").lower() != 'y':
            break
    except:
        print("Ошибка ввода. Попробуйте снова")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Упрощённые данные для Colab
cities = ["Москва", "Санкт-Петербург", "Казань", "Сочи", "Калининград"]
coords = {
    "Москва": (37.6176, 55.7558),
    "Санкт-Петербург": (30.3351, 59.9343),
    "Казань": (49.1088, 55.7961),
    "Сочи": (39.7233, 43.5855),
    "Калининград": (20.4542, 54.7104)
}

# Матрица расстояний (км)
dist_matrix = np.array([
    [0, 700, 800, 1600, 1250],
    [700, 0, 1500, 2300, 800],
    [800, 1500, 0, 1900, 2000],
    [1600, 2300, 1900, 0, 2850],
    [1250, 800, 2000, 2850, 0]
])

# Алгоритм Флойда-Уоршелла
def floyd_warshall(graph):
    n = len(graph)
    dist = graph.copy()
    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]
    return dist

shortest_dists = floyd_warshall(dist_matrix)

# Упрощённая визуализация без geopandas
def show_route(start_city, end_city):
    try:
        fig, ax = plt.subplots(figsize=(10, 6))

        # Рисуем все города
        for city, (lon, lat) in coords.items():
            ax.plot(lon, lat, 'o', markersize=8,
                   color='red' if city in [start_city, end_city] else 'blue')
            ax.text(lon+1, lat, city, fontsize=9)

        # Рисуем маршрут
        start_lon, start_lat = coords[start_city]
        end_lon, end_lat = coords[end_city]
        ax.plot([start_lon, end_lon], [start_lat, end_lat],
               'r--', linewidth=2)

        # Добавляем границы России (упрощённо)
        ax.plot([20, 180, 180, 20, 20], [40, 40, 75, 75, 40], 'k-', alpha=0.3)

        # Расстояние
        start_idx = cities.index(start_city)
        end_idx = cities.index(end_city)
        distance = shortest_dists[start_idx][end_idx]

        plt.title(f"{start_city} → {end_city} (~{distance} км)", fontsize=12)
        plt.xlabel("Долгота")
        plt.ylabel("Широта")
        plt.grid()
        plt.show()

    except Exception as e:
        print(f"Ошибка визуализации: {e}")
        start_idx = cities.index(start_city)
        end_idx = cities.index(end_city)
        print(f"Кратчайшее расстояние: {shortest_dists[start_idx][end_idx]} км")

# Интерфейс выбора
print("Доступные города:")
for i, city in enumerate(cities, 1):
    print(f"{i}. {city}")

while True:
    try:
        start_num = int(input("\nНомер начального города (1-5): ")) - 1
        end_num = int(input("Номер конечного города (1-5): ")) - 1

        if 0 <= start_num < 5 and 0 <= end_num < 5:
            show_route(cities[start_num], cities[end_num])
        else:
            print("Ошибка: введите числа от 1 до 5")

        if input("Продолжить? (y/n): ").lower() != 'y':
            break
    except:
        print("Ошибка ввода. Попробуйте снова")
```
